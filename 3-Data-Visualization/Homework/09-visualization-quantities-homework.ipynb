# Visualizing Quantities - Homework

Use the `birds.csv` dataset to practice plotting quantities.

```python
import pandas as pd
import matplotlib.pyplot as plt
birds = pd.read_csv('../../data/birds.csv')
birds.head()
```

Example: scatter plot of wingspan

```python
plt.figure(figsize=(10,4))
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.tick_params(axis='x',labelbottom=False)
for i in range(len(birds)):
    plt.plot(birds['Name'][i], birds['MaxWingspan'][i], 'bo')
plt.show()
```

Example: number of birds by category

```python
category_count = birds['Category'].value_counts()
category_count.plot.barh()
plt.title('Bird Categories')
plt.show()
```
