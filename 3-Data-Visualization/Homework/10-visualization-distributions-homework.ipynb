# Visualizing Distributions - Homework

Using the same `birds.csv` dataset, explore distributions.

```python
import pandas as pd
import matplotlib.pyplot as plt
birds=pd.read_csv('../data/birds.csv')
```

Example: histogram of `MaxBodyMass`.

```python
birds['MaxBodyMass'].plot(kind='hist',bins=30,figsize=(8,4))
plt.title('Distribution of Max Body Mass')
plt.show()
```

**Task:** Build a histogram showing the distribution of `MaxLength`.

Example: 2D histogram comparing mass and length.

```python
plt.hist2d(birds['MaxBodyMass'], birds['MaxLength'])
plt.xlabel('Max Body Mass')
plt.ylabel('Max Length')
plt.title('Mass vs Length')
plt.show()
```
