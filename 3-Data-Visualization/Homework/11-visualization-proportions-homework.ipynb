# Visualizing Proportions - Homework

We will use `mushrooms.csv` to practice proportion charts.

```python
import pandas as pd
import matplotlib.pyplot as plt
mushrooms=pd.read_csv('../data/mushrooms.csv')
```

Example: pie chart of poisonous vs edible classes.

```python
edibleclass=mushrooms.groupby('class').count()
labels=['Edible','Poisonous']
plt.pie(edibleclass['population'],labels=labels,autopct='%.1f %%')
plt.title('Edible?')
plt.show()
```

**Task:** Create a pie chart showing distribution of mushroom `odor`.


Example: donut chart of habitats.

```python
habitat=mushrooms.groupby('habitat').count()
labels=habitat.index.tolist()
plt.pie(habitat['class'],labels=labels,autopct='%1.1f%%',pctdistance=0.85)
center_circle=plt.Circle((0,0),0.40,fc='white')
fig=plt.gcf();fig.gca().add_artist(center_circle)
plt.title('Mushroom Habitats')
plt.show()
```

**Task:** Display a donut chart for `ring-type` values.

Example: waffle chart of cap colors.

```python
from pywaffle import Waffle
capcolor=mushrooms.groupby('cap-color').count()
data={'color':capcolor.index.tolist(),'amount':capcolor['class']}
df=pd.DataFrame(data)
fig=plt.figure(FigureClass=Waffle,rows=10,values=df.amount,labels=list(df.color),figsize=(12,8))
plt.show()
```
