# Creative Visualizations - Homework

Explore `letters.json` from the Dangerous Liaisons project and visualize the correspondence network.

```python
import json
import networkx as nx
import matplotlib.pyplot as plt
with open('../13-meaningful-visualizations/starter/src/assets/letters.json') as f:
    letters=json.load(f)
G=nx.DiGraph()
for l in letters:
    G.add_edge(l['from'], l['to'])
plt.figure(figsize=(8,8))
nx.draw_networkx(G, with_labels=True, node_size=500, font_size=8)
plt.show()
```

**Task:** Adjust node size based on how many letters each character sends.
